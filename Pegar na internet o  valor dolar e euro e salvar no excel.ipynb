```python
#importar a biblioteca selenium para trabalhar com as páginas web
from selenium import webdriver as opcoesSelenium
from selenium.webdriver.common.keys import Keys

#importando pyautogui para trabalhar com tempo e teclas
import pyautogui as tempoPausaPC
import pyautogui as teclasTeclado

#usando by para trabalhar com atualizações 
from selenium.webdriver.common.by import By

#passamos autorização de acesso as configurações do chrome
meuNavegador = opcoesSelenium.Chrome()
meuNavegador.get('https://www.google.com/')

#tempo para o pc processar as informações 
tempoPausaPC.sleep(2)

#pesquisar pelo elemento name ='q' ('barra de pesquisa') e após escrever no elemento a pesquisa 
meuNavegador.find_element(By.NAME, 'q').send_keys('Dolar hoje')

#tempo para o pc processar as informações 
tempoPausaPC.sleep(2)

#retorna o valor encontrado no campo name 'q' e faz a busca
meuNavegador.find_element(By.NAME, 'q').send_keys(Keys.RETURN)

#tempo para o pc processar as informações 
tempoPausaPC.sleep(2)

#No resultado pegamos o XPATH e no meios e depois pegando o primeiro elemento da lista
valorDolarPeloGoogle = meuNavegador.find_elements(By.XPATH, '//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]')[0].text

print(valorDolarPeloGoogle)

#---------------------------------------------------------------------------------------------- 

#tempo para o pc processar as informações 
tempoPausaPC.sleep(2)

#retorna o valor encontrado no campo name 'q' e faz a busca
meuNavegador.find_element(By.NAME, 'q').send_keys('')

#tempo para o pc processar as informações 
tempoPausaPC.sleep(2)

#estamos usando o pyautogui para apertar a tecla tab
teclasTeclado.press('tab')

#tempo para o pc processar as informações 
tempoPausaPC.sleep(2)

#estamos usando o pyautogui para apertar a tecla enter('para limpar o campode pesquisa')
teclasTeclado.press('enter')

#tempo para o pc processar as informações 
tempoPausaPC.sleep(2)

#pesquisar pelo elemento name ='q' ('barra de pesquisa') e após escrever no elemento a pesquisa 
meuNavegador.find_element(By.NAME, 'q').send_keys('Euro hoje')

#tempo para o pc processar as informações 
tempoPausaPC.sleep(2)

#retorna o valor encontrado no campo name 'q' e faz a busca
meuNavegador.find_element(By.NAME, 'q').send_keys(Keys.RETURN)

#tempo para o pc processar as informações 
tempoPausaPC.sleep(2)

#No resultado pegamos o XPATH e no meios e depois pegando o primeiro elemento da lista
valorEuroPeloGoogle = meuNavegador.find_elements(By.XPATH, '//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]')[0].text

#tempo para o pc processar as informações 
tempoPausaPC.sleep(2)

print(valorEuroPeloGoogle)

#-----------------------------------------------------------------------------------

#biblioteca para trabalhar com excel
import xlsxwriter
#biblioteca para abrir o excel
import os 

#salvar o caminho do arquivo e nomear o arquivo com a extensão desejada
caminhoArquivo = 'C:\\Users\\ADMIN\\Documents\\Python RPA e Excel como automatizar Processos e Planilhas!\SECAO 3 - ROBO 2 - EXTRAINDO DOLAR E EURO DA INTERNET\\Imprime Dolar e Euro.xlsx'

planilhaCriada = xlsxwriter.Workbook(caminhoArquivo)
sheet1 = planilhaCriada.add_worksheet()

#escrevendo nas células
sheet1.write('A1', 'Dólar')
sheet1.write('B1', 'Euro')

#valor em formato texto
sheet1.write('A2', valorDolarPeloGoogle)
sheet1.write('B2', valorEuroPeloGoogle)

#substituir a virgula por ponto, se não substituir vai dar erro no conversão para float
valorDolarPeloGoogle = valorDolarPeloGoogle.replace(',','.')
valorEuroPeloGoogle = valorEuroPeloGoogle.replace(',','.')

#transformar em float
dolar_float = float(valorDolarPeloGoogle)
euro_float = float(valorEuroPeloGoogle)

#célula float
sheet1.write('A3', dolar_float)
sheet1.write('B3', euro_float)


#Fechando arquivo do excel que está em segundo plano
planilhaCriada.close()

#abrir o arquivo
os.startfile(caminhoArquivo)
```
